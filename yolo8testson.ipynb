```python
import cv2
from ultralytics import YOLO
import os

# =========================
# AYARLAR
# =========================
MODEL_PATH = r"C:\Users\AhmetYasinUgur\Desktop\dataset\runs_yolov8\exp1\weights\best.pt"
OUT_DIR = r"C:\Users\AhmetYasinUgur\Desktop\dataset\runs\detect\webcam_live"

CONF = 0.45
IOU = 0.50
IMGSZ = 640
CAMERA_INDEX = 0        # 0 olmazsa 1 veya 2 dene
MIRROR = False          # True = ayna, False = normal

os.makedirs(OUT_DIR, exist_ok=True)

# =========================
# MODEL
# =========================
model = YOLO(MODEL_PATH)

# =========================
# KAMERA
# =========================
cap = cv2.VideoCapture(CAMERA_INDEX)

if not cap.isOpened():
    print("❌ Kamera açılamadı")
    exit()

width  = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps    = cap.get(cv2.CAP_PROP_FPS)
fps    = fps if fps > 0 else 20

# =========================
# VIDEO WRITER (KAYIT)
# =========================
out_path = os.path.join(OUT_DIR, "webcam_live.mp4")
fourcc = cv2.VideoWriter_fourcc(*"mp4v")
writer = cv2.VideoWriter(out_path, fourcc, fps, (width, height))

print("🎥 Canlı test başladı")
print("📂 Kayıt alınıyor:", out_path)
print("❌ Çıkmak için Q tuşuna bas")

# =========================
# CANLI DÖNGÜ
# =========================
while True:
    ret, frame = cap.read()
    if not ret:
        break

    if MIRROR:
        frame = cv2.flip(frame, 1)

    results = model.predict(
        frame,
        conf=CONF,
        iou=IOU,
        imgsz=IMGSZ,
        verbose=False
    )

    annotated = results[0].plot()

    writer.write(annotated)
    cv2.imshow("YOLO Webcam Live (Q to quit)", annotated)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# =========================
# TEMİZ KAPANIŞ
# =========================
cap.release()
writer.release()
cv2.destroyAllWindows()

print("✅ Canlı test kapandı")
print("💾 Video kaydedildi:", out_path)
```
